```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://en.wikipedia.org/wiki/List_of_largest_companies_in_India'
```

```python
x = requests.get(url)
y = BeautifulSoup(x.text,'html')
```

```python
header = y.find_all('table')[0]
```

```python
new = header.find_all('th')
```

```python
list_in = [ title.text.strip() for title in new]
```

```python
print(list_in)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = list_in)
df
```

```python
values = header.find_all('td')
list_in_values = [ title.text.strip() for title in values]
columns = list_in
num_columns = len(columns)
all_data = [list_in_values[i:i + num_columns] for i in range(0, len(list_in_values), num_columns)]
df = pd.DataFrame(all_data, columns=columns)
df
```

```python
df.to_csv('D:/Excel Macros/output2.csv', index=False)
```
